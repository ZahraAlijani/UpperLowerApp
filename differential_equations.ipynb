```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the differential equation
def f(x, y):
    return -2 * x * y

# Define the initial conditions
y0_lower = 0.9
y0_upper = 1.1

# Define the interval and step size
a, b = 0, 1
n = 10
h = (b - a) / n
x = np.linspace(a, b, n+1)

# Initialize arrays to store the solutions
y_lower = np.zeros(n+1)
y_upper = np.zeros(n+1)
y_euler = np.zeros(n+1)

# Set the initial conditions
y_lower[0] = y0_lower
y_upper[0] = y0_upper
y_euler[0] = (y0_lower + y0_upper) / 2

# Implement the fuzzy method (lower and upper functions)
for i in range(n):
    x_mid = x[i] + h / 2
    y_mid_lower = y_lower[i] + (h / 2) * f(x[i], y_lower[i])
    y_mid_upper = y_upper[i] + (h / 2) * f(x[i], y_upper[i])

    y_lower[i+1] = y_lower[i] + h * f(x_mid, y_mid_lower)
    y_upper[i+1] = y_upper[i] + h * f(x_mid, y_mid_upper)

    # Euler's method
    y_euler[i+1] = y_euler[i] + h * f(x[i], y_euler[i])

# Plot the results

plt.figure(figsize=(10, 6))
plt.plot(x, y_lower, label='Lower Bound (Fuzzy Method)', linestyle='--')
plt.plot(x, y_upper, label='Upper Bound (Fuzzy Method)', linestyle='--')
plt.plot(x, y_euler, label="Euler's Method", linestyle=':')
plt.fill_between(x, y_lower, y_upper, color='gray', alpha=0.3, label='Fuzzy Solution Range')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of Fuzzy Method and Euler\'s Method')
plt.legend()
plt.grid(True)
plt.margins(0)
plt.show()

# Create a table to compare the results
import pandas as pd

data = {
    'x': x,
    'Lower Bound (Fuzzy Method)': y_lower,
    'Upper Bound (Fuzzy Method)': y_upper,
    "Euler's Method": y_euler
}

df = pd.DataFrame(data)
df
```

```python
# Define the exact solutions for the corrected example
def exact_solution_lower(t):
    return 0.985 * np.exp(-t) - 0.025 * np.exp(t)

def exact_solution_upper(t):
    return 0.025 * np.exp(t) + 0.985 * np.exp(-t)

def exact_solution_mid(t):
    return np.exp(-t)

# Define the differential equation
f = lambda t, y: -y  # For both lower and upper approximations

# Parameters
h = 0.1  # Step size
t_start = 0  # Start time
t_end = 1  # End time
n = int((t_end - t_start) / h) + 1  # Number of time steps
t = np.linspace(t_start, t_end, n)  # Time points

# Initial conditions (trapezoidal fuzzy values)
Y_lower = np.zeros(n)
Y_upper = np.zeros(n)
Y_mid = np.zeros(n)

# Trapezoidal fuzzy initial values
y_initial = (0.96, 1.0, 1.0, 1.01)  # (a, b, c, d)
Y_lower[0] = y_initial[0]  # Lower bound (a)
Y_mid[0] = (y_initial[1] + y_initial[2]) / 2  # Midpoint of the core (average of b and c)
Y_upper[0] = y_initial[3]  # Upper bound (d)

# Apply the midpoint method for lower and upper approximations
for k in range(n - 1):
    t_mid = t[k] + h / 2

    y_lower_mid = Y_lower[k] + (h / 2) * f(t[k], Y_lower[k])
    Y_lower[k + 1] = Y_lower[k] + h * f(t_mid, y_lower_mid)

    y_upper_mid = Y_upper[k] + (h / 2) * f(t[k], Y_upper[k])
    Y_upper[k + 1] = Y_upper[k] + h * f(t_mid, y_upper_mid)

    y_mid_mid = Y_mid[k] + (h / 2) * f(t[k], Y_mid[k])
    Y_mid[k + 1] = Y_mid[k] + h * f(t_mid, y_mid_mid)

# Exact solutions for comparison
Y_exact_lower = exact_solution_lower(t)
Y_exact_upper = exact_solution_upper(t)
Y_exact_mid = exact_solution_mid(t)

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot approximations
plt.plot(t, Y_lower, linestyle='--', color='red', label="Lower Approximation")
plt.plot(t, Y_upper, linestyle='--', color='green', label="Upper Approximation")
plt.plot(t, Y_mid, linestyle='--', color='blue', label="Mid Approximation")

# Plot exact solutions
plt.plot(t, Y_exact_lower, color='orange', label="Exact Lower Solution")
plt.plot(t, Y_exact_upper, color='purple', label="Exact Upper Solution")
plt.plot(t, Y_exact_mid, color='black', label="Exact Mid Solution")

# Add labels and legend
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Lower, Upper, and Mid Approximations for Trapezoidal Fuzzy Initial Value Problem")
plt.grid(True)
plt.legend()
plt.xlim(0, 1)

# Show plot

# Save plot as PDF
plt.savefig("FIVP2.pdf", format='pdf')

# Show plot
plt.show()

```

```python
# Define the second-order Runge-Kutta (Heun's method)
def runge_kutta_2(f, y0, t):
    y = np.zeros(len(t))
    y[0] = y0
    h = t[1] - t[0]
    for i in range(1, len(t)):
        k1 = f(y[i-1], t[i-1])
        k2 = f(y[i-1] + h*k1, t[i-1] + h)
        y[i] = y[i-1] + (h/2) * (k1 + k2)
    return y

# Define the functions for the differential equations
def f1(y, x): return x**2 - y  # y' = x^2 - y
def f2(y, x): return 2*x - 8*x*np.cos(x**2)  # y' = 2x - 8x*cos(x^2)

# Define exact solutions
def exact_solution_1(t):
    return t**2 - 2*t + 2 - np.exp(-t)  # Analytical solution for y' = x^2 - y

def exact_solution_2(t):
    return 4 * np.sin(t**2) + 1  # Hypothetical analytical solution

# Define Euler's method
def euler_method(f, y0, t):
    y = np.zeros(len(t))
    y[0] = y0
    h = t[1] - t[0]
    for i in range(1, len(t)):
        y[i] = y[i-1] + h * f(y[i-1], t[i-1])
    return y

# Define time intervals for both differential equations
t1 = np.linspace(0, 2, 100)  # Interval for first differential equation
t2 = np.linspace(0, 5, 100)  # Interval for second differential equation

# Compute exact solutions
y_exact_1 = exact_solution_1(t1)
y_exact_2 = exact_solution_2(t2)

# Compute numerical solutions using different methods
y_euler_1 = euler_method(f1, 1, t1)
y_rk_1 = runge_kutta_2(f1, 1, t1)

y_euler_2 = euler_method(f2, 0, t2)
y_rk_2 = runge_kutta_2(f2, 0, t2)

# Compute errors
error_euler_1 = np.abs(y_exact_1 - y_euler_1)
error_rk_1 = np.abs(y_exact_1 - y_rk_1)
error_avg_1 = (error_euler_1 + error_rk_1) / 2

error_euler_2 = np.abs(y_exact_2 - y_euler_2)
error_rk_2 = np.abs(y_exact_2 - y_rk_2)
error_avg_2 = (error_euler_2 + error_rk_2) / 2

# Plot error comparison
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# First plot (0  t  2)
axs[0].plot(t1, error_rk_1, label='Second-order Runge-Kutta', color='darkblue')
axs[0].plot(t1, error_euler_1, label='Euler Method', color='green')
axs[0].plot(t1, error_avg_1, label='Average of Errors', color='black', linestyle='dashed')
axs[0].set_xlabel('t')
axs[0].set_ylabel('Error')
axs[0].set_xlim([0, 2])
axs[0].grid(True)
axs[0].legend()

# Second plot (0  t  5)
axs[1].plot(t2, error_rk_2, label='Second-order Runge-Kutta', color='darkblue')
axs[1].plot(t2, error_euler_2, label='Euler Method', color='green')
axs[1].plot(t2, error_avg_2, label='Average of Errors', color='black', linestyle='dashed')
axs[1].set_xlabel('t')
axs[1].set_ylabel('Error')
axs[1].set_xlim([0, 5])
axs[1].grid(True)
axs[1].legend()


# Show plot
fig.tight_layout()
fig.show()
fig.savefig("ivp4.pdf", format='pdf')
#pdf_filename = "/mnt/data/error_comparison.pdf"

```
